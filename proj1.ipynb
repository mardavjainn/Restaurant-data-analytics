```python
# Correlation matrix
# Determining if there is any correlation between the restaurant's location and its rating.

# Selecting relevant columns
df_corr = restaurant_rating[['Latitude', 'Longitude', 'Aggregate rating']]

# Computing correlation
correlation_matrix = df_corr.corr()
print(correlation_matrix)
```

```python
# Heatmap for Visualizing Correlation.

plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Location & Ratings")
plt.show()
```

```python
# Grouping by city/locality to check average ratings.
city_avg_ratings = restaurant_rating.groupby("City")["Aggregate rating"].mean().sort_values(ascending=False)
print(city_avg_ratings.head(10))  # Top 10 cities with highest average ratings
```

## Task 3

### Task: Feature Engineering

### Extract additional features from the existing columns, such as the length of the restaurant name or address.

```python
# Understanding data
restaurant_rating.head(5)
```

```python
# Adding extra features such as length of restaurant name or address.
restaurant_rating['length_of_restaurant_name'] = restaurant_rating['Restaurant Name'].str.len()
restaurant_rating['length_of_restaurant_address'] = restaurant_rating['Address'].str.len()
restaurant_rating.head(5)
```

###  Create new features like "Has Table Booking" or "Has Online Delivery" by encoding categorical variables

```python
# Creating new features by encoding categorical variables
restaurant_rating['Has Table booking'] = restaurant_rating['Has Table booking'].map({'Yes' : 1, 'No' : 0})
restaurant_rating['Has Online delivery'] = restaurant_rating['Has Online delivery'].map({'Yes' : 1, 'No' : 0})
restaurant_rating['Is delivering now'] = restaurant_rating['Is delivering now'].map({'Yes' : 1, 'No' : 0})
restaurant_rating.head(5)
```

```python
# Columns to be dropped
columns_to_drop = ['Restaurant ID', 'Restaurant Name', 'Address', 'Locality',
                   'Locality Verbose', 'Switch to order menu', 'Rating color',
                   'length_of_restaurant_name', 'length_of_restaurant_address']

# Dropping rating color column also because rating color and rating text columns are almost same.

restaurant_rating.drop(columns = columns_to_drop, inplace = True)
restaurant_rating.head()
```

```python
# Analyzing rating text and aggregate rating columns
restaurant_rating.groupby('Rating text')['Aggregate rating'].describe()
```

```python
# Performing encoding on Rating text column by manually labelling them
rating_text = {
    'Not rated' : -1,
    'Poor' : 0,
    'Average' : 1,
    'Good' : 2,
    'Very Good' : 3,
    'Excellent' : 4
}
restaurant_rating['Rating text'] = restaurant_rating['Rating text'].map(rating_text)
restaurant_rating.head(5)
```

```python
# Analyzing currency column
restaurant_rating['Currency'].value_counts()
```

```python
# One-hot encoding on currency column
restaurant_rating = pd.get_dummies(restaurant_rating, columns = ['Currency'], drop_first = True)
restaurant_rating.head()
```

```python
# Analyzing Country Code column
restaurant_rating['Country Code'].value_counts()
```

```python
# Since approximately 90% of the data belongs to country code 1, we create a new binary column, Is main country,
# to indicate whether a restaurant is in the primary country(1) or not(0). This allows us to retain the relevant information
# while dropping the Country Code column to reduce redundancy and improve model efficiency.

restaurant_rating['Is main country'] = restaurant_rating['Country Code'].apply(lambda x: 1 if x == 1 else 0)
restaurant_rating.drop("Country Code",axis = 1, inplace = True)
restaurant_rating.head()
```

```python
# Analyzing City column
restaurant_rating['City'].value_counts()
```

```python
# Grouping cities into major & other categories
top_cities = ['New Delhi', 'Gurgaon', 'Noida', 'Faridabad', 'Ghaziabad']
restaurant_rating['city_grouped'] = restaurant_rating['City'].apply(lambda x: x if x in top_cities else 'Other')
restaurant_rating.drop("City", axis=1, inplace = True)
restaurant_rating.head()
```

```python
# one hot encoding on "city_grouped column"
restaurant_rating = pd.get_dummies(restaurant_rating, columns = ['city_grouped'])
restaurant_rating.head()
```
